```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Task 1: Bernoulli Simulation

```python
def bernoulli_simulation(observation, run): # 100 runs, 10 observations -- 100 rows
    sample_mean = []
    for i in range(run):
        value = np.random.binomial(n=1, p=0.7, size=observation)
        sample_mean.append(np.mean(value))
    df = pd.DataFrame(sample_mean)
    file_name = str(observation) + '_obs_' + str(run) + '_runs.csv'
    df.to_csv(file_name, header=False, index=False)
    return sample_mean
```

```python
obs_list = [10, 100, 1000, 10000]
runs_list = [10, 100, 1000, 10000]
for obs in obs_list:
    for run in runs_list:
        dist = bernoulli_simulation(obs, run)
        plt.figure()
        plt.hist(dist, bins=20)
        plt.title(str(obs) + ' Observations and ' + str(run) + ' Runs')
        plt.xlabel('Sample Mean')
        plt.ylabel('Frequency')
```

Task 2: Analysis of HW1_dataset1

```python
df_set_1 = pd.read_csv('HW1_dataset_1.csv')
poisson_dist = np.random.poisson(lam=6, size=1000)

plt.title('Histogram of Dataset 1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.hist(df_set_1, bins='auto')
plt.hist(poisson_dist, bins='auto', alpha=0.5, label='Poisson Distribution')
plt.legend()
```

*Commentary:*

I chose the Poisson distribution to overlay on top of the dataset. Given that the shape of the distribution was slightly skewed, and started with 0 as the lowest value, I chose the Poisson distribution over the normal distribution. While the normal distribution also somewhat fit the shape of the data, it was more symmetrical due to the bell-shape of the curve.

Task 3: Analysis of HW1_dataset2

```python
df_set_2 = pd.read_csv('HW1_dataset_2.csv')
normal_dist = np.random.binomial(n=70, p=0.9, size=1000)

plt.title('Histogram of Dataset 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.hist(df_set_2, bins='auto')
plt.hist(normal_dist, bins='auto', alpha=0.5, label='Normal Distribution')
plt.legend()
```

*Commentary:*

Given that the shape of the dataset was more symmetrical, I chose a binomial distribution with an n-value of 70, and p-value of 0.9. This can be seen with the orange overlay on top of the dataset.

```python
df_set_3 = pd.read_csv('HW1_dataset_3.csv')

slope, intercept = np.polyfit(df_set_3['X'], df_set_3['Y'], 1)
line = slope * df_set_3['X'] + intercept

plt.plot(df_set_3['X'], line, color='red', label='Line of Best Fit')

plt.title('Scatterplot of Dataset 3')
plt.scatter(df_set_3['X'], df_set_3['Y'], alpha=0.2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
